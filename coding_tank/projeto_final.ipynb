```python

# colocar limitação se os pinos acabarem
# colocar .lower e .replace ( )


# lista = ['hugo', 'valeria', 'fernanda', 'paulo']
# 'valeria' in lista

print ('MASTERMIND')
import random

senha = random.sample(['vermelho', 'amarelo', 'roxo', 'verde', 'laranja', 'azul'], k=4)

print ('\n A senha foi definida \n')

print (senha)
# total de pinos no início e criação de variáveis
vermelho = 10
amarelo = 10
roxo = 10
verde = 10
laranja = 10
azul = 10
vermelho_proxy = 10
amarelo_proxy= 10
roxo_proxy = 10
verde_proxy = 10
laranja_proxy = 10
azul_proxy = 10
rodada = 1
pretos_brancos = []

    # while das rodadas. Coloco uma a mais para poder definir condições de derrota e vitória
while rodada <= 11:
        
#         condições de vitória e derrota
    if rodada == 11:
        print ('Você perdeu por exceder o número de chutes ):')
        break
    elif pretos_brancos == ['branco', 'branco', 'branco', 'branco']:
        print ('PARABÉNSSSS!!!! VOCÊ GANHOU!!')
        break
    else:
        pretos_brancos = []
        
    #     informa as rodadas e quantos pinos sobraram
        print (f'\nVocê tem {11 - rodada} chutes disponíveis')
        print (f'Você tem {vermelho} pinos vermelhos disponíveis')
        print (f'Você tem {amarelo} pinos amarelos disponíveis')
        print (f'Você tem {roxo} pinos roxos disponíveis')
        print (f'Você tem {verde} pinos verdes disponíveis')
        print (f'Você tem {laranja} pinos laranjas disponíveis')
        print (f'Você tem {azul} pinos azuis disponíveis\n')

    #  pede o chute ao jogador   
        resposta = input('Qual o seu chute? (escreva as cores em minúsculas e separadas por vírgula, sem espaço)').lower().replace(' ','').split(',')
        
        def cores_certas (resposta_para):
            vermelho_proxy = vermelho
            amarelo_proxy = amarelo
            roxo_proxy = roxo
            verde_proxy = verde
            laranja_proxy = laranja
            azul_proxy = azul
            
            for index_cor_3 in range(4):
                if resposta_para[index_cor_3] == 'vermelho':
                    if vermelho_proxy <= 0:
                        print ('Todos os pinos vermelhos já utilizados. Favor inserir a resposta de novo')
                        resposta = input('Qual o seu chute? (escreva as cores em minúsculas e separadas por vírgula, sem espaço)').lower().replace(' ','').split(',')
                        cores_certas(resposta)
                    elif vermelho_proxy > 0:
                        vermelho_proxy -= 1
                elif resposta_para[index_cor_3] == 'amarelo':
                    if amarelo_proxy <= 0:
                        print ('Todos os pinos amarelos já utilizados. Favor inserir a resposta de novo')
                        resposta = input('Qual o seu chute? (escreva as cores em minúsculas e separadas por vírgula, sem espaço)').lower().replace(' ','').split(',')
                        cores_certas(resposta)
                    elif amarelo_proxy > 0:
                        amarelo_proxy -= 1
                elif resposta_para[index_cor_3] == 'roxo':
                    if roxo_proxy <= 0:
                        print ('Todos os pinos roxos já utilizados. Favor inserir a resposta de novo')
                        resposta = input('Qual o seu chute? (escreva as cores em minúsculas e separadas por vírgula, sem espaço)').lower().replace(' ','').split(',')
                        cores_certas(resposta)
                    elif roxo_proxy > 0:
                        roxo_proxy -= 1
                elif resposta_para[index_cor_3] == 'verde':
                    if verde_proxy <= 0:
                        print ('Todos os pinos verdes já utilizados. Favor inserir a resposta de novo')
                        resposta = input('Qual o seu chute? (escreva as cores em minúsculas e separadas por vírgula, sem espaço)').lower().replace(' ','').split(',')
                        cores_certas(resposta)
                    elif verde_proxy > 0:
                        verde_proxy -= 1
                elif resposta_para[index_cor_3] == 'laranja':
                    if laranja_proxy <= 0:
                        print ('Todos os pinos laranjas já utilizados. Favor inserir a resposta de novo')
                        resposta = input('Qual o seu chute? (escreva as cores em minúsculas e separadas por vírgula, sem espaço)').lower().replace(' ','').split(',')
                        cores_certas(resposta)
                    elif laranja_proxy > 0:
                        laranja_proxy -= 1
                elif resposta_para[index_cor_3] == 'azul':
                    if azul_proxy <= 0:
                        print ('Todos os pinos vermelho já utilizados. Favor inserir a resposta de novo')
                        resposta = input('Qual o seu chute? (escreva as cores em minúsculas e separadas por vírgula, sem espaço)').lower().replace(' ','').split(',')
                        cores_certas(resposta)
                    elif azul_proxy > 0:
                        azul_proxy -= 1
                else:
                    print (f'{resposta_para[index_cor_3]} não existe na lista de cores liberadas')
                    resposta = input('Qual o seu chute? (escreva as cores em minúsculas e separadas por vírgula, sem espaço)').lower().replace(' ','').split(',')
                    cores_certas(resposta)
                return resposta_para
            
        resposta_definitiva = cores_certas(resposta)   #fornece a resposta tratada. Se der errado, cores_certas roda de novo até uma resposta condizente     
                        
                        

    #     cria a lista de pretos e brancos
        for index_cor in range(4):
            if resposta_definitiva[index_cor] == senha[index_cor]:
                pretos_brancos.append('branco')
            elif resposta[index_cor] == senha[0]:
                pretos_brancos.append('preto')
            elif resposta[index_cor] == senha[1]:
                pretos_brancos.append('preto')
            elif resposta[index_cor] == senha[2]:
                pretos_brancos.append('preto')
            elif resposta[index_cor] == senha[3]:
                pretos_brancos.append('preto')
            else:
                pretos_brancos.append('vazio')

    #     contabiliza as cores já usadas
        for index_cor_2 in range(4):
            if resposta_definitiva[index_cor_2] == 'vermelho':
                vermelho -= 1
            elif resposta_definitiva[index_cor_2] == 'amarelo':
                amarelo -= 1
            elif resposta_definitiva[index_cor_2] == 'roxo':
                roxo -= 1
            elif resposta_definitiva[index_cor_2] == 'verde':
                verde -= 1
            elif resposta_definitiva[index_cor_2] == 'laranja':
                laranja -= 1
            elif resposta_definitiva[index_cor_2] == 'azul':
                azul -=1
                
#      para avaliação do jogador de certos e errados
        print(f'\nSua resposta foi {resposta_definitiva}')
        print(f'Seus acertos são {pretos_brancos}')
        
        rodada += 1
        
            
            

```
