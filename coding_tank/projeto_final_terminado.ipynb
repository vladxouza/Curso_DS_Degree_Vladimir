```python

print('MASTERMIND')
import random

senha = random.sample(['vermelho', 'amarelo', 'roxo', 'verde', 'laranja', 'azul'], k=4)
cores_liberadas = ('vermelho', 'amarelo', 'roxo', 'verde', 'laranja', 'azul')
print('\n A senha foi definida \n')
print(senha)

# total de pinos no início e criação de variáveis
vermelho = 10
amarelo = 10
roxo = 10
verde = 10
laranja = 10
azul = 10
rodada = 1
pretos_brancos = []
resposta = []
# while das rodadas. Coloco uma a mais para poder definir condições de derrota e vitória
while rodada <= 11:

    #         condições de vitória e derrota
    if rodada == 11:
        print('Você perdeu por exceder o número de chutes ):')
        break
    elif pretos_brancos == ['branco', 'branco', 'branco', 'branco']:
        print('PARABÉNSSSS!!!! VOCÊ GANHOU!!')
        break
    else:
        pretos_brancos = []

    #     informa as rodadas e quantos pinos sobraram
    print(f'\nVocê tem {11 - rodada} chutes disponíveis')
    print(f'\nVocê tem {vermelho} pinos vermelhos disponíveis')
    print(f'Você tem {amarelo} pinos amarelos disponíveis')
    print(f'Você tem {roxo} pinos roxos disponíveis')
    print(f'Você tem {verde} pinos verdes disponíveis')
    print(f'Você tem {laranja} pinos laranjas disponíveis')
    print(f'Você tem {azul} pinos azuis disponíveis\n')

    #  pede o chute ao jogador e verifica erros na resposta e uso de pinos

    flag = True
    while flag:
        resposta = input(
            'Qual o seu chute? (escreva as cores em minúsculas e separadas por vírgula, sem espaço)').lower().replace(
            ' ', '').split(',')
        if len(senha) != len(resposta):
            print('A senha tem 4 cores diferentes. Inserir uma tentativa com 4 cores inclusas na lista')
            print(f'As cores liberadas são {cores_liberadas}')
            continue
        elif resposta[0] not in cores_liberadas or resposta[1] not in cores_liberadas or resposta[
                2] not in cores_liberadas or resposta[3] not in cores_liberadas:

            print('Uma das cores usadas não está na lista')
            print(f'As cores liberadas são {cores_liberadas}')
            continue
        vermelho_proxy = vermelho
        amarelo_proxy = amarelo
        roxo_proxy = roxo
        verde_proxy = verde
        laranja_proxy = laranja
        azul_proxy = azul
        index_cor_3 = 0
        flag_2 = True
        while flag_2:
            if index_cor_3 == 4:
                flag_2 = False
                break
            elif index_cor_3 < 4:
                if resposta[index_cor_3] == 'vermelho':
                    if vermelho_proxy <= 0:
                        print('Todos os pinos vermelhos já utilizados. Favor inserir a resposta de novo')
                        break
                    elif vermelho_proxy > 0:
                        vermelho_proxy -= 1
                elif resposta[index_cor_3] == 'amarelo':
                    if amarelo_proxy <= 0:
                        print('Todos os pinos amarelos já utilizados. Favor inserir a resposta de novo')
                        break
                    elif amarelo_proxy > 0:
                        amarelo_proxy -= 1
                elif resposta[index_cor_3] == 'roxo':
                    if roxo_proxy <= 0:
                        print('Todos os pinos roxos já utilizados. Favor inserir a resposta de novo')
                        break
                    elif roxo_proxy > 0:
                        roxo_proxy -= 1
                elif resposta[index_cor_3] == 'verde':
                    if verde_proxy <= 0:
                        print('Todos os pinos verdes já utilizados. Favor inserir a resposta de novo')
                        break
                    elif verde_proxy > 0:
                        verde_proxy -= 1
                elif resposta[index_cor_3] == 'laranja':
                    if laranja_proxy <= 0:
                        print('Todos os pinos laranjas já utilizados. Favor inserir a resposta de novo')
                        break
                    elif laranja_proxy > 0:
                        laranja_proxy -= 1
                elif resposta[index_cor_3] == 'azul':
                    if azul_proxy <= 0:
                        print('Todos os pinos vermelho já utilizados. Favor inserir a resposta de novo')
                        break
                    elif azul_proxy > 0:
                        azul_proxy -= 1
                index_cor_3 += 1
        if flag_2 == False:
            break
        else:
            continue

    #     cria a lista de pretos e brancos
    for index_cor in range(4):
        if resposta[index_cor] == senha[index_cor]:
            pretos_brancos.append('branco')
        elif resposta[index_cor] == senha[0]:
            pretos_brancos.append('preto')
        elif resposta[index_cor] == senha[1]:
            pretos_brancos.append('preto')
        elif resposta[index_cor] == senha[2]:
            pretos_brancos.append('preto')
        elif resposta[index_cor] == senha[3]:
            pretos_brancos.append('preto')
        else:
            pretos_brancos.append('vazio')

    #     contabiliza as cores já usadas
    for index_cor_2 in range(4):
        if resposta[index_cor_2] == 'vermelho':
            vermelho -= 1
        elif resposta[index_cor_2] == 'amarelo':
            amarelo -= 1
        elif resposta[index_cor_2] == 'roxo':
            roxo -= 1
        elif resposta[index_cor_2] == 'verde':
            verde -= 1
        elif resposta[index_cor_2] == 'laranja':
            laranja -= 1
        elif resposta[index_cor_2] == 'azul':
            azul -= 1

    #      para avaliação do jogador de certos e errados
    print(f'\nSua resposta foi {resposta}')
    print(f'Seus acertos são {pretos_brancos}')

    rodada += 1

            
            

```

```python
# --- MASTERMIND --- #

import random

print (" --- MASTERMIND --- \n")
print ("Guess the secret color code in as few tries as possible.\n")
print ("Please, enter your color code.\nYou can use red(R), green(G), blue(B), yellow(Y), white(W) and pink(P)")

colors = ["R", "G", "B", "Y", "W", "P"]
attempts = 0
game = True

# computer randomly picks four-color code
color_code = random.sample(colors,4)	
print (color_code)

# player guesses the number	
while game:
    correct_color = ""
    guessed_color = ""
    player_guess = input().upper()
    attempts += 1

    # checking if player's input is correct
    if len(player_guess) != len(color_code):
        print ("\nThe secret code has exactly four colors. I know, you can count to four. Try again!")
        continue
    for i in range(4):
        if player_guess[i] not in colors:
            print ("\nLook up what colors you can use in this game. You are not a daltonist, are you?")
            continue
            
    # comparison between player's input and secret code
    if correct_color != "XXXX":
        for i in range(4):
            if player_guess[i] == color_code[i]:
                correct_color += "X"
            if  player_guess[i] != color_code[i] and player_guess[i] in color_code:
                guessed_color += "O"
        print (correct_color +  guessed_color + "\n")		

    if correct_color == "XXXX":
        if attempts == 1:
            print ("Wow! You guessed at the first attempt!")
        else:
            print ("Well done... You needed " + str(attempts) + " attempts to guess.")
        game = False

    if attempts >= 1 and attempts <6 and correct_color != "XXXX":
        print ("Next attempt: ")
    elif attempts >= 6:
        print ("You didn't guess! The secret color code was: " + str(color_code))	

# play or not to play
    while game == False: 
        finish_game = input("\nDo you want to play again (Y/N)?").upper()	
        attempts = 0
        if finish_game =="N":
            print ("Thanks for the game! Bye, bye!")
        elif finish_game == "Y":
            game = True
            print ("So, let's play again... Guess the secret code: ")

# --- end --- #			
```
