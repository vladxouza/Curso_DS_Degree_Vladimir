    Área da esfera = 4*3.14*r*r;

```python
class Funcionario:
    def __init__(self,nome,email):
        '''
        Construtor
        
        Parâmetros:
        ----------
        nome : str
            O nome do funcionário
        email : str
            O email do funcionário
        dicionario_1 : dict
            Dicionário genérico que irá armazenar os meses como chaves e o valor como as horas trabalhadas no respectivo mês
        dicionario_2 : dict
            Dicionário genérico que irá armazenar os meses como chaves e o valor da hora no respectivo mês
        dicionario_3 : dict
            Dicionário genérico que irá armazenar o salário mensal calculado por cada mês
        '''
        self.nome=nome
        self.email=email
        self.dicionario_mes_horas={}
        self.dicionario_horas_valor={}
        self.dicionario_mes_valor={}
    def __repr__(self): #Método adicionado pelo exercício 13
        return f"Funcionario: {self.nome} - Email: {self.email}" 
    def adiciona_dados_salario(self,mes, horas, valor):
        '''
        Método que salva as informações digitadas pelo usuario nos respectivos dicionarios
        
        Parâmetros:
        ----------
        mes : str
            O mês a ser cadastrado
        horas : int
            A quantidade de horas trabalhadas no respectivo mês
        valor : float
            O valor da hora no mês em questão
        '''
        self.dicionario_mes_horas.update({mes:horas})
        self.dicionario_horas_valor.update({mes:valor})
    def calcula_salario(self):
        '''
        Método que calcula o salário do mês e o adiciona no dicionário de salarios mensais
        '''
        for mes in self.dicionario_mes_horas:
            self.dicionario_mes_valor.update({mes:(self.dicionario_mes_horas[mes]*self.dicionario_horas_valor[mes])}) 
    def mostra_salario(self):
        for mes in self.dicionario_mes_valor:
            print(f"Mês: {mes} - Salario: {self.dicionario_mes_valor[mes]}")

joao=Funcionario(nome='João',email='joao123@gmail.com')
joao.adiciona_dados_salario("Janeiro",220,20)
joao.adiciona_dados_salario("Fevereiro",220,45)
joao.calcula_salario()
joao.mostra_salario()
       
```

```python
import time

class Timer:
    def __init__(self,seconds):
        '''
        Constructor

        Parameters
        ----------
        seconds : int
            The seconds the class will work with
        '''
        self.seconds=seconds
    def __repr__(self): #Method added due to question 13
        return f"Time scheduled: {self.seconds} seconds"
    def countdown(self):
        for i in range (self.seconds,0,-1):
            print (i)
            time.sleep(1)
        print ("TIME IS OVER")
        
new_clock=Timer(10)
new_clock.countdown()
```

```python
class Phonebook:
    contacts={}
    
    def __init__(self,name,phone):
        '''
        Constructor

        Parameters
        ----------
        name : str
            The contact's name
        phone : str
            The contact's phone number
        '''
        self.name=name
        self.phone=phone
        Phonebook.contacts.update({self:[self.name,self.phone]})

    def add_contact(name,phone):
        '''
        A method that just calls the constructor and create a new instance, the parameters are the same
        '''
        return Phonebook(name,phone)

    @staticmethod
    def remove_contact(name):
        '''
        A method that looks for the desired contact on the dict above and delete it

        Parameters
        ----------
        name : str
            The name of the contact to delete
        '''
        for contact in Phonebook.contacts:
            if name==Phonebook.contacts[contact][0]:
                del Phonebook.contacts[contact]
                print("Contact Removed!")
                break

    @staticmethod
    def search_contact(name):
        '''
        A method that looks for the desired contact on the dict above and returns it

        Parameters
        ----------
        name : str
            The name you want to search for
        '''
        print("Results:")
        for contact in Phonebook.contacts:
            if name==Phonebook.contacts[contact][0]:
                print(f"Name: {Phonebook.contacts[contact][0]} | Phone: {Phonebook.contacts[contact][1]}")
    @staticmethod
    def print_contacts():
        '''
        A method that lists all the stored contacts
        '''
        for contact in Phonebook.contacts:
            print(f"Name: {Phonebook.contacts[contact][0]} | Phone: {Phonebook.contacts[contact][1]}")

Phonebook.add_contact("Yudi","40028922")

Phonebook.add_contact("Carlos","32334826")

Phonebook.add_contact("Luis","988241824")

Phonebook.print_contacts()

Phonebook.search_contact("Luis")

Phonebook.remove_contact("Luis")

Phonebook.print_contacts()
```

11 - Crie uma classe Fração cujos atributos são numerador (número de cima) e denominador (número de baixo).

Implemente os métodos de adição, subtração, multiplicação, divisão que retornam objetos do tipo Fração.

Implemente também o método _ repr _.

Implemente métodos para comparação: igualdade (==) e desigualdades (!=, <=, >=, < e >).

```python
class Fraction:
    def __init__(self, numerator, denominator):
        '''
        Constructor

        Parameters
        ----------
        numerator : int
            The numerator of the fraction
        denominator : int
            The denominator of the fraction

        Output : The methor will raise an error if the denominator is zero        
        '''
        if(denominator == 0):
            raise ValueError("The denominator can't be 0")
        self.numerator = numerator
        self.denominator = denominator
        
    def result(self):
        return self.numerator/self.denominator
        
    def __repr__(self):
        return f'{self.numerator}/{self.denominator}'
    
    def __add__(self, other):
        num = self.numerator*other.denominator +\
        self.denominator*other.numerator    
        den = self.denominator*other.denominator      
        return Fraction(num, den)
    
    def __sub__(self, other):
        num = self.numerator*other.denominator -\
        self.denominator*other.numerator
        den = self.denominator*other.denominator
        return Fraction(num, den)
    
    def __mul__(self, other):
        num = self.numerator * other.numerator
        den = self.denominator * other.denominator 
        return Fraction(num, den)
    
    def __truediv__(self, other):
        num = self.numerator*other.denominator
        den = self.denominator*other.numerator
        return Fraction(num, den)
    
    def __eq__(self, other):
        return self.result() == other.result()
    
    def __lt__(self, other):
        return self.result() < other.result()
    
    def __le__(self, other):
        return self.result() <= other.result()
    
    def __gt__(self, other):
        return self.result() > other.result()  
    
    def __ge__(self, other):
        return self.result() >= other.result()
```

```python
class Date:
    def __init__(self,day,month,year):
        '''
        Constructor

        Parameters:
        ----------
        month : int
            The month
        day : int
            The day
        year : int
            The year 

        Output: The init will raise errors if any of the entries isn't integer or out of range for a possible date.           
        '''
        if (not isinstance(month,int)) or ((month<1) or (month>12)): 
            raise ValueError("Month must be an integer between 1 and 12")
        if (not isinstance(day,int)) or ((day<1) or (day>31)):  
            raise ValueError("Day must be an integer between 1 and 31")   
        if (not isinstance(year,int)) or (day<0):
            raise ValueError("Year must be a positive integer")  
        self.month=month
        self.day=day
        self.year=year

    def __repr__(self):
        return f"{self.day:02d}/{self.month:02d}/{self.year}"

    def formatted_date(self):
        return(f"{self.year}{self.month}{self.day}")

    def __eq__(self, other):
        return self.formatted_date()==other.formatted_date()
    
    def __lt__(self, other):
        return self.formatted_date()<other.formatted_date()
    
    def __le__(self, other):
        return self.formatted_date()<=other.formatted_date()
    
    def __gt__(self, other):
        return self.formatted_date()>other.formatted_date()
    
    def __ge__(self, other):
        return self.formatted_date()>=other.formatted_date()

date1=Date(2,10,2016)
date2=Date(2,11,2011)
date3=Date(2,11,2011)

date2==date3
```

```python
class Customer:
    def __init__(self,name,age):
        '''
        Constructor

        Parameters
        ----------
        name : str
            The customer's name
        age : int
            The customer's age
        account : object
            The customer's account, object of the class account that will be bounded in it's init
        '''
        self.name=name
        self.age=age

class Account:
    def __init__(self,customer,balance):
        '''
        Constructor

        Parameters
        ----------
        customer : object
            The owner of the account, object of the class Customer
        balance : float
            The account's current balance

        '''
        self.customer=customer
        self.__balance=balance

    def deposit(self,money):
        '''
        Method to deposit money on the account

        Parameters
        ----------
        money : float
            The total amount to be added
        '''
        self.__balance+=money

    def __check_transaction(self,money):
        '''
        Method to check if the transaction is possible
        '''
        if money>self.__balance:
            raise ValueError ("Insufficient Funds!")

    def transfer(self,amount,destination_account):
        '''
        Method to transfer money from one account to another

        Parameters
        ----------
        amount : float
            The total amount to be transfered
        destination_account : object
            The account of destination (from class Account or VIPAccount)
        '''
        self.__check_transaction(amount)
        self.__balance-=amount
        destination_account.__balance+=amount
        print(f"New balance: {self.__balance}")

    def withdraw(self, amount):
        '''
        Method to withdraw money from the account

        Parameters
        ----------
        amount : float
            The total amount to be withdrawed
        '''
        self.__check_transaction(amount)
        self.__balance-=amount
        print(f"New balance: {self.__balance}")

    @property
    def balance(self):
        return self.__balance

class VIPAccount(Account):
    def __init__(self,customer,balance,overdraft):
        '''
        Constructor

        Parameters
        ----------
        customer : object, inherited from the class Account

        balance : float, inherited from the class

        overdraft : float
            The total amount avaiable to overdraft
        '''
        super().__init__(customer,balance)
        self.__overdraft=overdraft

    def __check_transaction(self,money):
        if money>(self.balance + self.__overdraft): 
            raise ValueError ("Insufficient Funds!")

    def withdraw(self, amount):
        self.__check_transaction(amount)
        if (self.balance-amount)>0:
            self.deposit(-amount)
        else:
            self.__overdraft+=(self.balance - amount)
            self.deposit(-self.balance)
            
        print(f"New balance: {self.balance}")
        print(f"Overdraft available: {self.__overdraft}")
    
    def transfer(self,amount,destination_account):
        self.__check_transaction(amount)
        if (self.balance-amount)>0:
            self.deposit(-amount)
        else:
            self.__overdraft+=(self.balance - amount)
            self.deposit(-self.balance)
        destination_account.deposit(amount)
        print(f"New balance: {self.balance}")
        print(f"Overdraft available: {self.__overdraft}")
    
    @property
    def overdraft(self):
        return self.__overdraft


customer01 = Customer("Leonardo",25)
B10001 = VIPAccount(customer01,1000,100)

customer02 = Customer("Yumi",27)
A10002 = Account(customer02,0)
```

```python
class Rectangle:
    def __init__(self,a_side,b_side):
        '''
        Constructor
        
        Parameters:
        ----------
        a_side : float
            One side of the rectangle
        b_side : float
            The other side of the rectangle
        '''
        self.a_side=a_side
        self.b_side=b_side
    def __repr__(self): #Method added due to question 13
        return f"Side One:{self.a_side} Side Two:{self.b_side} "
    def total_area(self):
        area=(self.a_side*self.b_side)
        return area   
class Square(Rectangle):
    def __init__(self,a_side,b_side):
        super().__init__(a_side,b_side)
        if a_side != b_side:
            raise TypeError ("Object is not a Square")        

example_rectangle=Rectangle(4,6)
example_square=Square(4,4)
not_a_square=Square(4,6) #This line will cause a crash
```
